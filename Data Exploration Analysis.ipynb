# Popularity

There is a large chunk of the data that is scored very lowly. Which could be damaging to generalble model. We need to make it a bit simpler for the model and generalizable. 

```python
df['popularity'].hist(bins=50)
```

```python
#lets convert to between 0 - 9 for scores. 
```

```python
score = []
for v in df['popularity']:
    if v >=50:
        score.append('popular')
    else:
        score.append('unpopular')
```

```python
df['score'] = score
```

```python
df.head()
```

## Train a simple Random Forest
The goal of this notebook is find feature importance to reduce the number of features to use and make the model as generalizable as possible. Removing too highly coalated data as well to the score. So will train a model on all available data and then simplify


```python
label = 'score'
cat_features = ['year', 'mode','genre_label', 'time_signature', 'key']
num_features = ['danceability', 'energy',  'loudness', 'speechiness',
       'acousticness', 'instrumentalness', 'liveness', 'valence', 'tempo',
       ]
```

```python
X = df[cat_features + num_features]
y = df[label]
```

```python
X.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
model_X = scaler.fit(X)
X_scaled = model_X.transform(X)

X_train, X_test, y_train, y_test = train_test_split (X_scaled, y, random_state=42, test_size=0.3, shuffle=True)
```

```python
len(X_train), len(X_test)
```

```python
#importing necessary libraries to perform logistic regression
from sklearn import datasets, linear_model, metrics

log_reg = linear_model.LogisticRegression(solver = 'lbfgs', max_iter = 1000)

#train the model using training set
log_reg.fit(X_train, y_train)

#making predictions on the testing set
y_pred = log_reg.predict(X_test)

#printing accuracy of algorithm
print("Logistic Regression Model Accuracy (in %):",
metrics.accuracy_score(y_test, y_pred) * 100)
```

# Train Random forest

```python
#importing confusion matrix and classification report
from sklearn.metrics import classification_report, confusion_matrix
```

```python
rf = RandomForestClassifier(n_jobs=8)
rf.fit(X_train, y_train)
y_pred = rf.predict(X_test)

#printing confusion matrix and classification report
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

```python
rf.feature_importances_
```

```python
# Extract feature importances
feature_importances = rf.feature_importances_

# Create a DataFrame for visualization
feature_importance_df = pd.DataFrame({
    'Feature': X.columns,
    'Importance': feature_importances
})
```

```python
# Sort the DataFrame by importance
feature_importance_df = feature_importance_df.sort_values(by='Importance', ascending=False)

# Plot the feature importances
plt.figure(figsize=(10, 6))
sns.barplot(x='Importance', y='Feature', data=feature_importance_df, palette='viridis')
plt.title('Feature Importances')
plt.xlabel('Importance')
plt.ylabel('Feature')
plt.show()
```
